```python
# Importing necessary libraries
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time

# Set up the Selenium WebDriver (Make sure you have ChromeDriver installed)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# URL for Tesla revenue data
url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'

# Open the URL in the browser
driver.get(url)

# Wait for the page to load completely
time.sleep(5)  # Adjust sleep time if necessary

# Get the page source
html = driver.page_source

# Print the first 500 characters of the HTML content to inspect the structure
print("First 500 characters of the page source:")
print(html[:500])

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Try finding the table (print a message if it's not found)
table = soup.find('table')

if table:
    print("\nTable found. Extracting rows...")
    # Extract all rows from the table
    rows = table.find_all('tr')
    
    # Prepare list to hold revenue data
    revenue_data = []
    
    # Loop through rows, skipping the header row
    for row in rows[1:]:
        columns = row.find_all('td')
        if len(columns) > 1:  
            year = columns[0].text.strip()
            revenue = columns[1].text.strip()
            revenue_data.append([year, revenue])
    
    # Create DataFrame from the extracted data
    tesla_revenue = pd.DataFrame(revenue_data, columns=['Year', 'Revenue'])
    
    # Print the DataFrame to the notebook
    print("\nTesla Revenue Data:")
    print(tesla_revenue.tail())
else:
    print("No table found on the page.")

# Close the browser window
driver.quit()
```

```python
import yfinance as yf

# Extract GameStop stock data for the last 1 year
gme_data = yf.download('GME', period='1y')

# Reset the index
gme_data.reset_index(inplace=True)

# Save and display the first five rows of the gme_data dataframe
print("\nFirst 5 rows of GameStop stock data:")
print(gme_data.head())
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time

# Set up Selenium WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# URL for GameStop revenue data
url_gme = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'

# Open the page in the browser
driver.get(url_gme)

# Wait for the page to load completely
time.sleep(5) 

# Get the page source after it's fully loaded
html = driver.page_source

# Print the first 500 characters of the HTML content for debugging
print("\nFirst 500 characters of the GameStop page source:")
print(html[:500])

# Parse the HTML content with BeautifulSoup
soup_gme = BeautifulSoup(html, 'html.parser')

# Try finding the table (print a message if it's not found)
table_gme = soup_gme.find('table')

if table_gme:
    print("\nTable found. Extracting rows for GameStop revenue data...")
    # Extract the rows from the table
    rows_gme = table_gme.find_all('tr')

    # Prepare list to hold revenue data
    revenue_data_gme = []
    
    # Loop through rows, skipping the header row
    for row in rows_gme[1:]:
        columns = row.find_all('td')
        if len(columns) > 1:  
            year = columns[0].text.strip()
            revenue = columns[1].text.strip()
            revenue_data_gme.append([year, revenue])

    # Create DataFrame from the extracted data
    gme_revenue = pd.DataFrame(revenue_data_gme, columns=['Year', 'Revenue'])
    
    # Print the DataFrame to the notebook
    print("\nGameStop Revenue Data:")
    print(gme_revenue.tail())
else:
    print("No table found on the page.")

# Close the browser window
driver.quit()
```

```python
import matplotlib.pyplot as plt

# Function to plot stock data
def make_graph(stock_data, title):
    stock_data['Close'].plot(figsize=(10, 6))
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.grid(True)
    plt.show()

# Plot Tesla stock data
print("\nPlotting Tesla Stock Price...")
make_graph(tesla_data, 'Tesla Stock Price')
```

```python
# Plot GameStop stock data
print("\nPlotting GameStop Stock Price...")
make_graph(gme_data, 'GameStop Stock Price')
```
